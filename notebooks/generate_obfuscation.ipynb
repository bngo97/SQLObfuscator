```python
from obfuscator import Obfuscator
import pandas as pd
import numpy as np
import re
```

```python
def clean_string(query):
    return re.sub('\s+', ' ', query.replace('(', ' ').replace(')', ' ').replace(',', ' , ')).strip()
```

```python
df = pd.read_csv('../data/all_data.csv')
df['payload'] = df['payload'].apply(clean_string)
df.insert(2, 'is_obf', 0, True)
```

```python
obf = Obfuscator(1, 50)
row_list = []
# Generate obfuscation for each payload
for _, row in df.iterrows():
    payload = row['payload']
    obfuscated_payload = obf.obfuscate(payload)
    row_list.append({'payload': obfuscated_payload, 'length': len(obfuscated_payload), 'is_obf': 1, 'label': row['label']})
```

```python
# Create obfuscated dataframe
df_obfuscated = pd.DataFrame(row_list)
```

```python
# Save obfuscated data and non_obfuscated data
df_obfuscated.to_csv('obf_data.csv', index=None, header=True)
df_combined = pd.concat([df, df_obfuscated], ignore_index=True)
df_combined.to_csv('all_data.csv', index=None, header=True)
```
